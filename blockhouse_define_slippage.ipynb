# 📘 Define Slippage

Slippage quantifies the **deviation between the actual execution price** of a trade and the **expected (mid) price**. It arises due to **limited liquidity** in the limit order book (LOB) and grows as order size increases.

### 🧮 Formal Definition

The **temporary market impact function** $g_t(x)$ is defined as:

$$
g_t(x) = \text{Actual Price} - \text{Expected Price} = \frac{\text{Cost}(x)}{x} - \text{Mid}_t = \frac{\sum_{i=0}^{k} p_i \cdot \min(q_i, r_i)}{x} - \frac{\text{best bid} + \text{best ask}}{2}
$$

---
### 🧩 Explanation of Terms:

- $x$: Size of the market order to buy  
- $p_i$: Ask price at LOB level $i$  
- $q_i$: Available quantity at ask level $i$  
- $r_i$: Remaining quantity to fill at that stage (initially $r_0 = x$)  
- $\text{Mid}_t$: The mid-price at time $t$ = $\frac{\text{best bid} + \text{best ask}}{2}$

---
### 💡 Intuition:

- For a **small** market order, you often fill at the top level — low slippage.  
- For a **large** order, you walk the book and consume multiple levels, increasing total cost and slippage.  
- This formula precisely models that effect by summing costs across available LOB levels until the order is filled.
